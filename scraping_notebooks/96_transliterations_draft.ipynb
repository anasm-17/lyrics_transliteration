```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://lyricstranslate.com/en/nǐ-zěnme-shuō-你怎么说-nǐ-zěnme-shuō.html"
```

```python
html_content = requests.get(url).text
soup = BeautifulSoup(html_content)
for i in soup.find_all(class_="ltf"):
    print(i.get_text() + "shhh")
```

```python
url2 = 'https://lyricstranslate.com/en/collection/chinese-songs-transliteration-easy-learning'
```

```python
html_content2 = requests.get(url2).text
```

```python
soup2 = BeautifulSoup(html_content2)
soup2
```

```python
links_list = []
for link in soup2.find_all("a",href = True):
    if link["href"].startswith("/en/") and link["href"].endswith("-lyrics.html"):
        print(link.get_text)
```

```python
links_list = []
for link in soup2.find_all("a",href = True):
        links_list.append(link["href"])
```

```python
links_list
```

```python
for i in soup2.find("span", class_="collection-title").children:
    print(i["href"])
```

```python
links_list = []
for i in soup2.find_all("span", class_="collection-title"):
    for c in i.children:
        links_list.append(c['href'])
```

```python
len(links_list)
```

```python

for link in links_list:
    html_content = requests.get("https://lyricstranslate.com" + link).text
    soup = BeautifulSoup(html_content)
    for i in soup.find_all(class_="ltf"):
        print(i.get_text())
```
